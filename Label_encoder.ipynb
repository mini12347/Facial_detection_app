```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import pickle


df = pd.read_csv('./data_emotion/data (1).csv')

# Get unique labels
labels = df['label'].values

# Create and fit label encoder
le = LabelEncoder()
le.fit(labels)

print("Label Encoder created!")
print(f"Classes: {le.classes_}")
print(f"Number of classes: {len(le.classes_)}")

# Save it
with open('label_encoder.pkl', 'wb') as f:
    pickle.dump(le, f)

print("\n✓ label_encoder.pkl saved successfully!")
print("You can now run the webcam script.")

# Test it
with open('label_encoder.pkl', 'rb') as f:
    le_test = pickle.load(f)
    
print("\nTest - encoding examples:")
for cls in le_test.classes_:
    encoded = le_test.transform([cls])[0]
    print(f"  {cls} -> {encoded}")
```
